```python
import pandas as pd
import numpy as np
import scipy.stats as st

import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(34)
```

```python
dist = st.t(12, 0, 1.5)
data = st.t(12, 0, 1.5).rvs(size=(2000, 100))
```

```python
pdf = dist.pdf(s := np.linspace(-8, 8, 1000))

with sns.plotting_context('poster'):
    
    plt.style.use('seaborn-whitegrid')
    
    fig, ax = plt.subplots(1, 1, figsize=(30, 15))
    sns.despine(fig, left=True, bottom=True)
    
    for row in data[:1250]:
        sns.kdeplot(row, ax=ax, lw=1, alpha=.15, color='yellow')
    for row in data[1250:]:
        sns.kdeplot(row, ax=ax, lw=1, alpha=.05, color='blue')
    
    ax.plot(s, pdf, lw=4, color='black', linestyle='dashed')
    
    ax2 = fig.add_axes([0.395, 0.16, 0.20, 0.20])
    im = plt.imread('Python_logo_icon-1345481447.png')
    ax2.imshow(im)
    ax2.axis('off')
    
    fig.savefig('logo.png', dpi=300, bbox_inches='tight')
```
